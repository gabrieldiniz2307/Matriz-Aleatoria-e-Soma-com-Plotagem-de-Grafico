```python
import numpy as np
import matplotlib.pyplot as plt
import random as randint
```

```python
linhas_matriz = 3
colunas_matriz = 3

matriz_a = np.random.randint(0, 30, size=(linhas_matriz, colunas_matriz))
matriz_b = np.random.randint(0, 30, size=(linhas_matriz, colunas_matriz))

print(matriz_a)
print(matriz_b)
```

```python
soma = matriz_a[1,0] + matriz_b[0,1] + matriz_a[2,0] + matriz_b[0,2] + matriz_a[2,1] + matriz_b[1,2]
print(soma)
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(matriz_a, cmap='viridis')
plt.colorbar()
plt.title('Gráfico da Matriz A')
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(matriz_b, cmap='viridis')
plt.colorbar()
plt.title('Gráfico da Matriz B')
plt.show()
```
